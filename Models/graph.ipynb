```python
import pandas as pd

# Load the Excel file
file_path = '../Data/E05_Hudson_South_West_10_min_avg_20220129_20230128.xls'
data = pd.read_excel(file_path)

# Display the first few rows of the dataframe to understand its structure and contents
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtering the dataset for columns related to wind speed
wind_speed_columns = [col for col in data.columns if 'HorizWS' in col or 'VertWs' in col]
wind_speed_data = data[wind_speed_columns]

# Basic statistics of the wind speed data
wind_speed_statistics = wind_speed_data.describe()

# Plotting the distributions of the wind speed data
plt.figure(figsize=(12, 6))
for col in wind_speed_columns:
    sns.distplot(wind_speed_data[col], label=col, hist=False)

plt.title('Distribution of Wind Speed Measurements')
plt.xlabel('Wind Speed')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

wind_speed_statistics
```
